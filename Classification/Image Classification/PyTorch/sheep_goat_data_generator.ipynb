```python
import os
import random
import csv
import shutil
import pandas as pd
import matplotlib.pyplot as plt
```

```python
sheep_path = r'../datasets/sheep_goat_classification/sheeps/'
goat_path = r'../datasets/sheep_goat_classification/goats/'
train_path = r'../datasets/sheep_goat_classification/train/'
test_path = r'../datasets/sheep_goat_classification/test/'
```

```python
data_list = []

sheep_files = os.listdir(sheep_path)
goat_files = os.listdir(goat_path)

for f in sheep_files:
    data_list.append([f,0])

for f in goat_files:
    data_list.append([f,1])

random.shuffle(data_list)
train_list = data_list[:-572]
test_list = data_list[-572:]

data_list.insert(0,['id','sheep_or_goat'])
train_list.insert(0,['id','sheep_or_goat'])
test_list.insert(0,['id','sheep_or_goat'])

print(data_list)
print(str(len(data_list)))
print(train_list)
print(str(len(train_list)))
print(test_list)
print(str(len(test_list)))
```

```python
with open('../datasets/sheep_goat_classification/data.csv','w', newline='') as file:
    writer = csv.writer(file)
    writer.writerows(data_list)

with open('../datasets/sheep_goat_classification/train.csv','w', newline='') as file:
    writer = csv.writer(file)
    writer.writerows(train_list)

with open('../datasets/sheep_goat_classification/test.csv','w', newline='') as file:
    writer = csv.writer(file)
    writer.writerows(test_list)
```

```python
train_data = pd.read_csv('../datasets/sheep_goat_classification/train.csv')
print('Label Distribution: \n' + str(train_data['sheep_or_goat'].value_counts()))
```

```python
label = 'Has Sheep', 'Has Goat'
plt.figure(figsize=(8, 8))
plt.pie(train_data.groupby('sheep_or_goat').size(), labels=label, autopct='%0.1f%%', startangle=90)
plt.show()
```

```python
for index, row in train_data.iterrows():
    if row['sheep_or_goat'] == 0:
        src = os.path.join(sheep_path,row['id'])
        if shutil.copy(src,train_path):
            print("File " + row['id'] + " copied from " + sheep_path + " to " + train_path + " successfully!!!")
    else:
        src = os.path.join(goat_path,row['id'])
        if shutil.copy(src,train_path):
            print("File " + row['id'] + " copied from " + goat_path + " to " + train_path + " successfully!!!")
```

```python
test_data = pd.read_csv('../datasets/sheep_goat_classification/test.csv')
print('Label Distribution: \n' + str(test_data['sheep_or_goat'].value_counts()))
```
